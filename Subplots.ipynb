Here are some examples of using the matplotlib plotting facility.

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
# z is sin function phased pi/2 (90 degrees - curve shifted to the left)
z = np.sin(x + (np.pi/2))
# w is sin functions phased -pi/2 (90 degrees - curve shifted to the right)
w = np.sin(x - (np.pi/2))
```

```python
# Create one single diagram as a subplot
fig, (ay, az, aw) = plt.subplots(1, 3, sharey=True)
ay.plot(x,y)
ay.set_title('Simple plot')
az.plot(x,z)
az.set_title('sin(x+pi/2)')
aw.plot(x,w)
aw.set_title('sin(x-pi/2)')
# fig.show
```

```python
# Create a subplot with 2 diagrams
# Notice the sharing of the y axis in both diagrams.
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Here is an example of polar plots.
# Notice that only 2 of the 4 plots in the figure are actually plotted;
# postion (0,0) and (1,1).
fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
axes[0, 0].plot(x, y)
axes[1, 1].scatter(x, y)
```

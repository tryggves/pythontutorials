```python
import numpy as np
import matplotlib.pyplot as plt

z = np.arange(1,10,0.1)
fz = np.power(z, 2)

print ("Hello world")
print("z:\n{}".format(z))
print("fz:\n{}".format(fz))

plt.plot(z, fz, marker='.')
```

```python
# Set up an array of 2 seconds in steps of 10ms
t=np.arange(0.0, 2.0, 0.01)
# Create a sine wave
s=1+np.sin(2*np.pi*t)
# Create another time series
s2=1+np.sin(4*np.pi*t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure() and then ax = fig.add_subplot(111)
# (ax1, ax2) is a tuple i.e. an immutable list.
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True)
ax1.plot(t, s)

ax1.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax1.grid()

ax2.plot(t, s2)
ax2.plot()

ax2.set(xlabel='time (s)', ylabel='voltage (mV)')
ax2.grid()


fig.savefig("test2.png")
plt.show()
```
